```python
import pandas as pd
import pickle

df = pd.read_csv('../data/test.csv')

df['datetime'] = pd.to_datetime(df['datetime'])
df['hour'] = df['datetime'].dt.hour
df['dayofweek'] = df['datetime'].dt.dayofweek
df['month'] = df['datetime'].dt.month
df['year'] = df['datetime'].dt.year

df.head(3)
```

```python
X_test = df[['datetime', 'season', 'holiday', 'workingday', 'weather', 'temp', 'atemp', 'humidity', 'windspeed', 'hour', 'dayofweek', 'month', 'year']]
X_test_new = X_test[['datetime', 'hour', 'month', 'year', 'holiday', 'season', 'weather', 'temp', 'humidity', 'windspeed', 'dayofweek']]
```

```python
# load the transformation from disk
fe = pickle.load(open('feature_2.pkl', 'rb'))
```

```python
Xtest_tran = fe.transform(X_test_new)
pd.DataFrame(Xtest_tran, columns=fe.get_feature_names_out())
```

```python
# load the model from disk
model = pickle.load(open('model_2.pkl', 'rb'))
```

```python
predictions = model.predict(Xtest_tran)
predictions
```

```python
X_test.head(3)
```

```python
import numpy as np
datetime = X_test.datetime
submission = pd.DataFrame({ 'datetime' : np.array(datetime), 'Count': np.maximum(np.array(predictions.round().astype(int)), 0) })
submission.to_csv('submission.csv', index=False)
```
